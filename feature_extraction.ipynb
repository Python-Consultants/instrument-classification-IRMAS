```python
def fileFolder2FeatureTable(instrument_type, path):
    rootpath = path +'/' +instrument_type
    files = os.listdir(rootpath)
    train_set_total = []
    train_label_total = []
    for i in files:
        y, sr = librosa.load(rootpath+'/'+i)
        m_slaney = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40, dct_type=2)
        #m_htk = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40, dct_type=3)
        #single_array = m_slaney.reshape(1,40,130)
        train_label_total.append(instrument_type)
        train_set_total.append(m_slaney)
        train_set_array = np.array(train_set_total)
    return (train_set_array, train_label_total)
```

```python
voi_raw, voi_label = fileFolder2FeatureTable('voi','IRMAS-TrainingData')
vio_raw, vio_label = fileFolder2FeatureTable('vio','IRMAS-TrainingData')
tru_raw, tru_label = fileFolder2FeatureTable('tru','IRMAS-TrainingData')
sax_raw, sax_label = fileFolder2FeatureTable('sax','IRMAS-TrainingData')
pia_raw, pia_label = fileFolder2FeatureTable('pia','IRMAS-TrainingData')
org_raw, org_label = fileFolder2FeatureTable('org','IRMAS-TrainingData')
gel_raw, gel_label = fileFolder2FeatureTable('gel','IRMAS-TrainingData')
gac_raw, gac_label = fileFolder2FeatureTable('gac','IRMAS-TrainingData')
flu_raw, flu_label = fileFolder2FeatureTable('flu','IRMAS-TrainingData')
cla_raw, cla_label = fileFolder2FeatureTable('cla','IRMAS-TrainingData')
cel_raw, cel_label = fileFolder2FeatureTable('cel','IRMAS-TrainingData')
```

```python
raw_data_all = np.concatenate([voi_raw, vio_raw, tru_raw, sax_raw, pia_raw, org_raw, gel_raw, gac_raw, flu_raw, cla_raw, cel_raw])
```

```python
#测试全连接网络
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(40, 130)),
    keras.layers.Dense(256)
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dense(11, activation=tf.nn.softmax)
])
model.compile(optimizer=tf.train.AdamOptimizer(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10)
```

```python
#测试CNN
from __future__ import print_function
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv1D
from keras.callbacks import ModelCheckpoint
from keras.models import model_from_json
from keras import backend as K

model = Sequential()
model.add(Conv1D(64,(1), input_shape=(40,130), activation='relu', padding="same"))
model.add(Conv1D(64,(1), activation='relu', padding="same"))
model.add(Conv1D(64,(1), activation='relu', padding="same"))
model.add(Flatten())
model.add(Dense(64, activation='softmax'))
model.add(Dense(11, activation='softmax'))

model.compile(optimizer = tf.train.AdamOptimizer(),
              loss='sparse_categorical_crossentropy', #keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
batch_size = 128
epochs = 10
model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv1D, MaxPooling1D
model = Sequential()
model.add(Conv1D(32, (5), padding='same',
                 input_shape=(40,130)))
model.add(Activation('relu'))
model.add(Conv1D(32, (5)))
model.add(Activation('relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.25))

model.add(Conv1D(64, (3), padding='same'))
model.add(Activation('relu'))
model.add(Conv1D(64, (3)))
model.add(Activation('relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(11))
model.add(Activation('softmax'))
model.compile(optimizer = tf.train.AdamOptimizer(),
              loss='sparse_categorical_crossentropy', #keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
batch_size = 128
epochs = 10
model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1)
```
